##                                      UNEMPLOYMENT ANALYSIS WITH PYTHON



```python
import pandas as pd
unemployment_data = pd.read_csv('Unemployment in India.csv')
print(unemployment_data.head())
unemployment_data.drop(['Area'], axis=1, inplace=True)
unemployment_data[' Date'] = pd.to_datetime(unemployment_data[' Date'])
unemployment_data.set_index(' Date', inplace=True)
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(unemployment_data[' Estimated Unemployment Rate (%)'], label='Unemployment Rate')
plt.xlabel(' Date')
plt.ylabel(' Unemployment Rate (%)')
plt.title(' Unemployment Rate in India over Time')
plt.legend()
plt.grid(True)
plt.show()
plt.figure(figsize=(12, 6))
plt.plot(unemployment_data[' Estimated Employed'], label='Employed', color='green')
plt.plot(unemployment_data[' Estimated Labour Participation Rate (%)'], label='Labour Participation Rate', color='blue')
plt.xlabel('Date')
plt.ylabel('Count/Rate')
plt.title('Employed and Labour Participation Rate in India over Time')
plt.legend()
plt.grid(True)
plt.show()
region_wise_unemployment = unemployment_data.groupby('Region')[' Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
region_wise_unemployment.plot(kind='bar', color='skyblue')
plt.xlabel('Region')
plt.ylabel('Mean Unemployment Rate (%)')
plt.title('Mean Unemployment Rate by Region')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
